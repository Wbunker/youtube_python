Think of yield as a "Just In Time" function that can generate a list.  

Each item in the list is generated by what is returned by the yield statement.

The normal way of creating lists return all the items at once.  If you have a large list this could take a lot of time or take up a lot of memory.


What if every item in the list was a car.  You wouldn't want to build and order 10,000 cars to sit in a lot waiting for people to come by and grab one.

Yield let's you build each car the minute someone needs it.


Here is the world's simplest generator.  It keeps track of how many times it has been called.

```python
def simple_generator():
    x = 1
    while True:
        yield x
        x += 1
```

We can call the generator, one of 2 ways.  
First is with next()

```python
number_in_queue = simple_generator()

# let's deal with first 10 customers
for i in range(10):
    print(f"You are customer {next(number_in_queue)} ")
```

We can also use it in for statement.
But in this case, we have to be careful because our generator doesn't stop

```python
new_customers = 0

for customer_number in number_in_queue:
    print(f"You are customer {next(number_in_queue)} ")
    new_customers += 1
    if new_customers > 20:
        break
```
